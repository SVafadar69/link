```python
import os 
from pathlib import Path 
import random 
import shutil

image_folder = os.path.join(os.getcwd(), 'package_images')

def download_filtered_images_locally(image_paths: list[str], output_dir: str) -> None: 
    output = Path(output_dir)
    output.mkdir(parents=True, exist_ok=True)

    for image_path in image_paths: 
        src = Path(image_path)
        if src.exists():
            dest = output / src.name 
            shutil.copy2(src, dest)
            print(f'Downloaded: {src.name}')
        else: 
            print(f'Skipped {image_path}')


def combine_images(base_dir: str, sample_size: int = 1000) -> list[str]: 
    base = Path(base_dir)
    image_extensions = {'.jpg', '.jpeg', '.png', '.tiff', '.webp'}
    splits = {'train', 'test', 'valid'}
    filtered_images = []

    for split in splits: 
        images_dir = base  / split / 'images'
        if images_dir.exists():
            for file in sorted(images_dir.iterdir()):
                if file.suffix.lower() in image_extensions: 
                    filtered_images.append(str(file))
    
    actual_sample_size = min(sample_size, len(filtered_images))
    selected_images = random.sample(filtered_images, actual_sample_size)
    return selected_images
    

#download_filtered_images_locally(combine_images(image_folder), 'package_images')
```

```python
### Put images under one dir 
def get_jpeg_images(folder: str) -> list[Path]:
    return [f for f in Path(folder).iterdir() if f.suffix.lower() == '.jpeg']


def clean_image_name(filename: str) -> str:
    # "1_jpg.rf.abc123.jpg" -> split on first "jpg" -> "1_" -> "1_.jpeg"
    base = filename.split('jpg')[0]  # "1_"
    return base + 'jpeg'


def copy_to_cleaned(images: list[Path], output_dir: Path) -> None:
    output_dir.mkdir(parents=True, exist_ok=True)

    for src in images:
        new_name = clean_image_name(src.name)
        dest = output_dir / new_name
        shutil.copy2(src, dest)
        print(f'Copied: {src.name} -> {new_name}')


def process_images(package_images_dir: str) -> None:
    folder = Path(package_images_dir)
    cleaned_dir = folder / 'cleaned_images'

    images = get_jpeg_images(str(folder))
    print(f'Found {len(images)} JPEG images')

    copy_to_cleaned(images, cleaned_dir)
    print(f'Done. Cleaned images saved to: {cleaned_dir}')


#process_images('package_images')
```

```python
import cv2
import numpy as np 
from typing import Tuple
from PIL import Image

def resize_with_letterbox(image_path, target_size: Tuple[int] = (640, 640), color: Tuple[int] = (114, 114, 114)): 
    image = Image.open(image_path)
    h, w = image.shape[:2]
    scale = min(target_size[0] / w, target_size[1] / h)
    new_w, new_h = int(w * scale), int(h * scale)

    resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    canvas = np.full((target_size[1], target_size[0], 3), color, dtype=np.uint8)

    x_offset = (target_size[0] - new_w) // 2 
    y_offset = (target_size[1] - new_h) // 2
     
    canvas[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = resized 
    return canvas

def display_image(image):
    Image.open(image)

display_image(resize_with_letterbox("/Users/sv/Desktop/Desktop - S’s MacBook Pro/link/model_training/deer_images/0_deer.jpg"))
```

```python
import os 
import shutil 

def combine_images_into_all(root_dir = os.path.join(os.getcwd(), "training_set")):
    extensions = (".png", ".jpg", ".jpeg", ".webp")
    all_images_dir = os.path.join(root_dir, "all_images")
    os.makedirs(all_images_dir, exist_ok=True)
    
    for folder in os.listdir(root_dir):
        folder_path = os.path.join(root_dir, folder)
        if (not os.path.isdir(folder_path) or folder == "all_images"):
            continue 
        for filename in os.listdir(folder_path):
            if filename.lower().endswith(extensions):
                src = os.path.join(folder_path, filename)
                dst = os.path.join(all_images_dir, filename)
                shutil.copy2(src, dst)

combine_images_into_all()
```
