Augmentation Best Practices 
- Get more real world data if model performance is low 
- Only apply augmentations to make model invariant to ___ condition 
- 1. Crop or resize first 2. Horizontal Flip, SquareSymmetry (if doing aerial/medical) 3. Dropout/Occlusion (CoarseDropout, ConstrainedCourseDropout -> very impactful) 4. Color dependence (convert to gray scale, channel dropout ((channel drop may/may not be useful)) 5. Affine transformations (scale/rotation) 6. Domain-specific transforms (for your use case only) 7. Normalizing (standard/sample-specific - do last) 

Training Ideas 
- Train multiple models on different augmentations - see progress of each (dashboard) + the model weights 
- Simulate models in the real world: evals for each augmentation/run (what are the evals for training vision models? How to actually determine my model is smart?)
- How is the accuracy of models actually tested in industry before deployment? 
- Get high-quality, real-world labelled data 
- Agent to autonomously collect internet data for you - labelled/unlabelled. 
- Train multiple models at once (re-use data): yoloe, yolov11

- training automation: auto parameters/pytorch paper
- automated bbox segmentation: sam + text prompt + masks -> bbox conversions (Nvidia Locate Anything)
- local auto-annotation: local models to choose from, preset, upload images + perform task (huggingface connected)
